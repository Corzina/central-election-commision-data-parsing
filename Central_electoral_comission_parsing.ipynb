```python
import pandas as pd
import numpy as np
from selenium import webdriver as wb
from bs4 import BeautifulSoup
```

В качестве примера мы парсим данные по выборам депутатов Законодательного Собрания Тверской области седьмого созыва

```python
br = wb.Chrome(r'C:\Chromedriver\chromedriver.exe') 
#укажите путь к chromedriver на вашем ПК

br.get('http://www.tver.vybory.izbirkom.ru/region/izbirkom?action=show&root=1&tvd=26920001674675&vrn=26920001674670&prver=2&pronetvd=null&region=69&sub_region=69&type=424&report_mode=null')
#укажите выборы на сайте ЦИКа, которые вы хотите спарсить. Специально не удаляю ссылку, чтобы можно было увидеть пример.
```

```python
table = BeautifulSoup(br.page_source).find_all('table', {'class':'table-fixed-columns table-bordered table-striped table-sm'})[0]
allstuff = []

for number in table.find_all('td', {'class':'text-right'}):
    voting = number.find('b').get_text()
    allstuff.append(voting)
allstuff

# К этому моменту у нас есть табличка, спарсенная в виде строки. 
# Осталось конвертировать её, создав пустую табличку и скачать. С этим придется немного поколдовать.
```

```python
len(allstuff) 

# И так, нам известно число элементов матрицы. В моем случае - это 1113. Я посмотрел делители числа 1113 и понял, что
# нам понадобится пустая матрица 21 на 53, чтобы конвертировать строку в таблицу.
```

```python
pd.DataFrame(index = [list(range(1,22))], columns = [list(range(1,54))]) 
# обратите внимание - мы создаем число строчек и столбцов на 1 больше
```

```python
# Заливаем строку в пустую таблицу
input = allstuff
num = np.array(input)
reshaped = num.reshape(21,53)
electoraldata = pd.DataFrame(reshaped)
electoraldata
```

```python
# теперь можно просто сохранить файл
electoraldata.to_excel('Antihype.xlsx')
```
